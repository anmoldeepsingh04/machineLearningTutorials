## Simple Linear Regression
We will use linear regression to predict the home prices, based on a given dataset.
We'll have a look at the dataset first.

```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.linear_model as linear_model # this library has all the models

df = pd.read_csv('homeprices.csv')
df
```

Firstly, we will plot the data and then we want to predict the prices for two homes with ares of 3300 sqft and 5000 sqft, respectively.

```python
plt.xlabel('Area in sqft')
plt.ylabel('Price in $')
plt.plot(df.area, df.price, 'rx')
plt.show
```

Now, we will create a line of best fit, which has the least distance from our datapoints. To find that line, we need to have the equation of the line, which will look like the following:



Here, `area` is the independent variable and `price` is called the dependent variable.

Now we will look at the code implementation of finding the line of best fit.

```python
reg = linear_model.LinearRegression()

# now we will fit the data, which means training the linear regression model
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3300]])
```

So, when we trained our model on the data, it calculated the coefficient and the slope of the line with the best fit.

```python
print(reg.coef_[0])
print(reg.intercept_)
```

So now, we will find the prices of houses whose area is given in `areas.csv` file and put the prices in the second columns.

```python
areaDataFrame = pd.read_csv('areas.csv')
prices = reg.predict(areaDataFrame)
areaDataFrame['prices'] = prices
areaDataFrame.to_csv('predictions.csv', index=False)
```

Now we will plot our data and the line of best fit in the same plot to get a feel of how things look.

```python
plt.title('Price Prediction')
plt.xlabel('Area in sqft')
plt.ylabel('Price in $')
plt.plot(df.area, df.price, 'rx')
plt.plot(df.area, reg.predict(df[['area']]))
plt.show()
```
