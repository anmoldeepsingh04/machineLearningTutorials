## What is machine learning?
There are certain tasks where computers really excel compared to humans like:
1. Computing the product of two large numbers
2. Search through millions of records to find the desired data

While, computers are great at certain tasks, there are plethora of tasks where humans outperform machines like:
1. Driving a car
2. Natural language conversation

So, machine learning is the process of making computers better at doing things that humans can do better than machines. We make machines learn things like humans do. For doing this, we will look at specific tasks and try to understand why humans are better than machines in those tasks. We can understand this by observing and understanding how the human brain works.

The human brain consists of tiny neurons that are present in billions inside the brain and whenever we think or make a decision, a chemical signal is generated and these neurons light up. When an input is provided to the brain, it connects a certain set of neurons, connected through synapses and creating a neural pathway to produce the desired output. Overtime, when the same input is provided, the same set of neurons light up and the same output is received, which strengthens those pathways, and the brain "learns" the information.

We can understand this through the following example.

Consider a human baby, when they see an object, the parent tells them that the object is a Cow or a Car, certain neurons in their brain light up and those neural pathways become stronger. When this process is repeated many times over a period, the baby learns what a Cow or a Car looks like.

 

The learning process includes the following steps:
1. Observing
2. Giving a name (Predict)
3. Making mistakes
4. Correcting the mistake
5. Learning again

During learning, the edges or pathways in the baby's brain are constantly adjusted and the new pathways are being created which leads to a certain output or decision.

Computers can be treated in a similar way where neural network models are created and trained with a lot of data. This is the subfield of ML which is known as Deep Learning.

So, machine learning in a broad sense can be defined as a set of techniques to extract knowledge from available data and using that knowledge to make decisions.

Machine learning has profound effects in our daily life, from autonomous cars to spam filtering in emails, personal assistance devices, recommedation systems on netflix or youtube, etc.
