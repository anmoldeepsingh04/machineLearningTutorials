We have data for the per capita income of Canada in US$ from 1970 to 2016 and we want to predict the per capita income for the year 2020.

```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.linear_model as linmod

df = pd.read_csv('canada_per_capita_income.csv')
df.head()

plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Training Data')
plt.plot(df['year'], df['per capita income (US$)'], 'rx')
```

```python
# training the model
linearRegression = linmod.LinearRegression()
linearRegression.fit(df[['year']].values, df['per capita income (US$)'].values)

# making prediction for 2020
print(linearRegression.predict([[2020]]))
```

```python
# printing the data and the line of best fit

plt.plot(df.year, df['per capita income (US$)'], 'rx', label = 'Training Datapoints')
plt.plot(df.year, linearRegression.predict(df[['year']].values), color='green', label = f'y = {linearRegression.coef_[0]:0.2f}x + {linearRegression.intercept_:0.2f}')
plt.plot(2020, linearRegression.predict([[2020]]), 'xb')
plt.vlines(2020, 0, linearRegression.predict([[2020]]))
plt.hlines(linearRegression.predict([[2020]]), 0, 2020 )
plt.xlim(1968, 2022)
plt.ylim(0, )
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.legend()
plt.show()
```
