# ML Tutorial Day 5

We can save trained models in files and use them as and when we need them. The training process of a machine learning model, where the training dataset is in gigabytes, if not bigger, is the most time-consuming part. Once a model is trained we can save it for later use and save a lot of time. We can ave the model using two methods:

## Pickle

1. using the pickle module
We can save the model as a `.pickle` file and use the pickle object to directly make predictions.

```python
import pandas as pd
from sklearn import linear_model as linMod
import pickle

df = pd.read_csv('homeprices.csv')
df
reg = linMod.LinearRegression()
reg.fit(df[['area']].values, df[['price']].values)
```

```python
print(reg.predict([[2000]]))

# saving the model as pickle file
with open('model_pickle', 'wb') as f:
    pickle.dump(reg, f)
```

```python
# loading the model
with open('model_pickle', 'rb') as f:
    mp = pickle.load(f)

# using model to make predictions
mp.predict([[2000]])
```

## Joblib

2. using joblib from sklearn. Usually, it is believed that if you have a large number of numpy-arrays, then using joblib is preferred as compared to pickle.

```python
import joblib

# saving the model as joblib file
joblib.dump(reg, 'model_joblib')

# loading the model
mj = joblib.load('model_joblib')
# using to make predictions
mj.predict([[2000]])
```
