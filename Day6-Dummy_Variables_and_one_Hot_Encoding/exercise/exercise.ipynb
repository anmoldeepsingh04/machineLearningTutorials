# Exercise

`carprices.csv` has car sell prices for 3 different models. First plot data points on a scatter plot chart to see if linear regression model can be applied. If yes, then build a model that can answer following questions,

1) Predict price of a mercedez benz that is 4 yr old with mileage 45000

2) Predict price of a BMW X5 that is 7 yr old with mileage 86000

3) What is the accuracy of your model

```python
import pandas as pd
from sklearn import linear_model as linMod
import matplotlib.pyplot as plt

# importing and visualizing data
df = pd.read_csv('carprices.csv')
df.head()
```

```python
plt.xlabel('Age(yrs)')
plt.ylabel('Mileage')
plt.plot(df['Age(yrs)'], df['Mileage'], 'rx')
plt.plot(df['Age(yrs)'], df['Sell Price($)'], 'g+')
plt.show()
```

```python
# creating one hot encoding using pandas

# creating dummy variables
dummies = pd.get_dummies(df['Car Model'])

# merging them with main data frame
merged = pd.concat([df, dummies], axis = 1)

# creating the final dataframe
final = merged.drop(['Car Model', 'Mercedez Benz C class'], axis=1)
final
```

```python
# preparing data for input
X = final.drop(['Sell Price($)'], axis = 1).values
Y = final['Sell Price($)'].values

# training the linear regression model
reg = linMod.LinearRegression()
reg.fit(X, Y)
```

```python
# making predictions

# 1) Predict price of a mercedez benz that is 4 yr old with mileage 45000

# 2) Predict price of a BMW X5 that is 7 yr old with mileage 86000

# 3) What is the accuracy of your model

print("The price of a mercedez benz that is 4 yr old with mileage 45000 is: {}".format(reg.predict([[45000, 4, 0, 0]])[0]))
print("The price of a BMW X5 that is 7 yr old with mileage 86000 is: {}".format(reg.predict([[86000, 7, 0, 1]])[0]))
print(f"The accuracy of the model is: {100*reg.score(X,Y)}%")
```
